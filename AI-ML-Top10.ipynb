Convert duration column to timedelta

```python
videos_df['duration'] = videos_df['duration'].apply(lambda x: timedelta(seconds=sum(int(i[:-1]) * {'H': 3600, 'M': 60, 'S': 1}[i[-1]] for i in re.findall(r'\d+[HMS]', x))))
```

Format the duration as HH:MM:SS

```python
videos_df['duration'] = videos_df['duration'].apply(lambda x: str(datetime.min + x).split(' ')[1])
```

## To Find out Top 10 YouTubers

*We will verify it by the Channel Subscribers & Average Views gained per video uploaded*

```python
channels_df.head()
```

```python
top_ten = channels_df.drop(columns = ['channel_id','description'])
```

```python
top_ten.head()
```

```python
top_ten['views_count_per_video'] = top_ten.view_count / top_ten.video_count
```

```python
top_ten.head()
```

### Current Top 10 By Subscriber Count

If we go by the count of subscribers of each of the channels, here are the Top 10 Youtube Channels currently. 

```python
top_ten_by_subs = top_ten.sort_values(by = ['subscriber_count'], ascending = False).head(10)
top_ten_by_subs
```

```python
top_ten_by_subs.channel_title
```

### Current Top 10 By Views per Video Uploaded

If we go by the number of views acchieved on an average per video uploaded, here are the Top 10 Youtube Channels currently.

```python
top_ten_by_avg = top_ten.sort_values(by = ['views_count_per_video'], ascending = False).head(10)
top_ten_by_avg
```
